```python
import numpy as np
from scipy.sparse import random
from scipy import stats
from numpy.random import default_rng
from matplotlib import pyplot as plt
from Graph import Graph, Painter
from utils import CustomDistribution
import os
```

# define the input, recurrent and output matrices of a small circuit

```python
rng = default_rng()
X = CustomDistribution(seed=rng)
Y = X() # get a frozen version of the distribution
N = 10
density = 3/N
W_rec = 0.8 * random(N, N, density=density, random_state=rng, data_rvs=Y.rvs).toarray()
np.fill_diagonal(W_rec, 0)
W_inp = np.zeros((N, 2))
W_inp[0, 0] = 1
W_inp[2, 1] = 1
W_inp = W_inp + 0.2 * np.random.randn(N, 2)

W_out = np.zeros((3, N))
W_out[:, N-3:] = np.eye(3) + 0.2 * np.random.randn(3, 3)
```

# define the graph, initialize nodes and edges

```python
G = Graph(W_inp=W_inp, W_rec=W_rec, W_out = W_out,
          inp_node_labels=["one", "two"],
          out_node_labels=["I", "II", "III"],
          rec_node_labels=['|' + str(i) + '|' for i in range(N)])
fig, ax = G.draw_graph()
plt.tight_layout()
plt.show()
```
